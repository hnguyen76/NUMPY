```python
from numpy import random
```

### Random module functions
- random.rand()
random.randn() generate a single random number
random.randn(n) generate an array of n random numbers
random.randn(m,n) genrate a 2d array with m row, n columns
random.choice()
random.shuffle() shuffle contents of array

```python
data = random.rand(10).round(4)*100
data1 = random.randn(10).round(4)*100
print(f'rand(): {data}')
print(f'randn(): {data1}')
```

```python
elements = ['a', 'b', 'c','d','e','f']
random_sample = random.choice(elements)
print(random_sample)
```

```python
random.shuffle(elements)
print(elements)
```

```python
menu = [
    "Spaghetti Carbonara",
    "Chicken Alfredo",
    "Margherita Pizza",
    "Cheeseburger",
    "Caesar Salad",
    "Fish and Chips",
    "Pad Thai",
    "Sushi Platter",
    "Vegetable Stir-Fry",
    "Grilled Salmon"
]
```

```python
def suggest_dish():
    random_dish = random.choice(menu)
    return random_dish
```

```python
print(suggest_dish())
```
