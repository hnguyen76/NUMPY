```python
one_to_100 = np.arange(1,100)
print(one_to_100)
```

```python
arr = np.arange(1,11)
print(arr)
```

```python
sum_arr = arr + 10
print(sum_arr)
```

```python
sub_arr = arr - 10
print(sub_arr)
```

```python
mult_arr = arr * 10
print(mult_arr)
```

```python
div_arr = arr/2
print(div_arr)
```

```python
sum =np.sum(arr)
mean = np.mean(arr)
median = np.median(arr)
standard_dev = np.std(arr).round(2)
min = np.min(arr)
max = np.max(arr)
variance = np.var(arr)

print(f'Sum: {sum}')
print(f'Mean: {mean}')
print(f'Median: {median}')
print(f'SD: {standard_dev}')
print(f'Min: {min}')
print(f'Max: {max}')
print(f'variance: {variance}')
```

```python
matrix = np.array([[1,2,3],[4,5,6]]) 
print(matrix)
```

```python
transposed_matrix = matrix.T
print(transposed_matrix)
```

```python
reshaped_matrix = matrix.reshape(3,2)
print(reshaped_matrix)
```

#Sorting, Searching and Counting

```python
print(np.argmin(arr))
print(np.argmax(arr))
```

```python
newArr = np.array([0,1,2,0,3,0,0,0,2])
print(np.count_nonzero(newArr))
print(np.count_nonzero(newArr)/len(newArr))
```

```python
fruitArr = np.array(['apple','banana','apple','apple','plum','mango','banana'])
print(np.unique(fruitArr))
```

```python
decade1 = np.arange(1980, 1990)
decade2 = np.arange(1990, 2000)
decade3 = np.arange(2000, 2010)

print(decade1)
print(decade2)
print(decade3)
allYears = np.concatenate((decade1,decade2))
print(allYears)
```
