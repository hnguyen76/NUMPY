### NumPy and Math

```python
import numpy as np
```

```python
python_list = [1,2,3,4,5]

numpy_array = np.array(python_list)
print(numpy_array)
```

```python
shape = numpy_array.shape
size = numpy_array.size
type = numpy_array.dtype

print(f'1D NumPy Array: {shape}')
print(f'number of Array: {size}')
print(f'type of Array: {type}')
```

```python
arr1 = np.array([10,20,30,40])
arr2 = np.array([2,4,6,8])

subtraction = arr1-arr2
print(subtraction)
```

```python
multiplication = arr1*arr2
print(multiplication)
print(multiplication.dtype)
```

```python
division = arr1/arr2
print(division)
print(division.dtype)
```

```python
data = np.array([85,90,78,92,88,85,95])

min_value = np.min(data)
max_value = np.max(data)
data_range = max_value - min_value
median = np.median(data)
std = np.std(data).round(4)
variance = np.var(data).round(4)

print(f'Min: {min_value}')
print(f'Median: {median}')
print(f'Std: {std}')
print(f'Variance: {variance}')
print(f'Range: {data_range}')
```

```python
temperature_data = np.array(
    [22.3, 23.1,
    24.5, 25.8, 23.6, 26.7,
    27.9, 29.2, 30.5, 24.7,
    23.4, 22.1, 25.3, 26.4,
    28.7, 29.8, 31.2, 32.4,
    30.7, 29.5, 27.8, 26.6,
    23.9, 22.5, 24.1, 25.7,
    27.3, 29.6, 31.0, 33.1,
    31.9
    ]
)
meantemp = np.mean(temperature_data)
median = np.median(temperature_data)
std = np.std(temperature_data)
```

```python
print(f'Mean: {meantemp}')
print(f'Median: {median}')
print(f'Standard Deviation: {std}')
```

```python
Ftemp = temperature_data*9/5 +32
print(Ftemp)
```

```python
hotdays = temperature_data[temperature_data>30]
print(f'Hot day: {hotdays}')
print(f'Num of hot days: {len(hotdays)}')
```

```python
base_temp = 20
cooling_degree_total = np.sum(np.maximum(temperature_data-base_temp,0))
print(f'Cooling degree total: {cooling_degree_total}')
```
