### NumPy and Math

```python
import numpy as np
```

```python
python_list = [1,2,3,4,5]

numpy_array = np.array(python_list)
print(numpy_array)
```

```python
shape = numpy_array.shape
size = numpy_array.size
type = numpy_array.dtype

print(f'1D NumPy Array: {shape}')
print(f'number of Array: {size}')
print(f'type of Array: {type}')
```

```python
arr1 = np.array([10,20,30,40])
arr2 = np.array([2,4,6,8])

subtraction = arr1-arr2
print(subtraction)
```

```python
multiplication = arr1*arr2
print(multiplication)
print(multiplication.dtype)
```

```python
division = arr1/arr2
print(division)
print(division.dtype)
```

```python
data = np.array([85,90,78,92,88,85,95])

min_value = np.min(data)
max_value = np.max(data)
data_range = max_value - min_value
median = np.median(data)
std = np.std(data).round(4)
variance = np.var(data).round(4)

print(f'Min: {min_value}')
print(f'Median: {median}')
print(f'Std: {std}')
print(f'Variance: {variance}')
print(f'Range: {data_range}')
```
